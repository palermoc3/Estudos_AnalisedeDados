### üìä <span style="color:#2E86C1;"><u>Criando um DataFrame com Pandas</u></span>

Aqui estamos usando o `pd.DataFrame` para criar uma **tabela estruturada**, que √© um dos principais recursos da biblioteca **Pandas**.  
Um DataFrame funciona como uma **planilha**, possuindo **linhas, colunas e r√≥tulos**, o que facilita a organiza√ß√£o e an√°lise dos dados.

üìå **No exemplo abaixo**, estamos montando uma **matriz de confus√£o**, que √© usada em **Machine Learning** para avaliar a performance de um modelo de classifica√ß√£o. Ela compara os valores **reais** com os valores **previstos**, permitindo identificar acertos e erros do modelo.

```python
df = pd.DataFrame(
    {
        'Valor Obtido: Verdadeiro': [990, 10],
        'Valor Obtido: Falso': [2, 998]
    },
    index=['Valor Real: Verdadeiro', 'Valor Real: Falso']
)


print(df)
```
