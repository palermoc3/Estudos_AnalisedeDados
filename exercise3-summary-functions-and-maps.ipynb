## 5.
I'm an economical wine buyer. Which wine is the "best bargain"? Create a variable `bargain_wine` with the title of the wine with the highest points-to-price ratio in the dataset.

```python
# 1. Calcular a relação custo-benefício (points/price)
# Adicionamos uma nova coluna (opcional, mas bom para a análise)
reviews['cost_benefit'] = reviews.points / reviews.price

# 2. Encontrar o rótulo do índice (linha) que tem o valor máximo na coluna 'cost_benefit'
best_value_index = reviews.cost_benefit.idxmax()

# 3. Usar o rótulo do índice encontrado para retornar a linha completa
bargain_wine = reviews.loc[best_value_index].title

bargain_wine
```
